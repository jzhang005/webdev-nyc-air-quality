```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import geopandas as gpd
```

```python
# import NYC area data from 'Neighborhood Tabulation Areas.geojson'
areas = gpd.read_file('Neighborhood Tabulation Areas.geojson')
areas.columns
```

```python
# visualize nyc tabulation
areas.plot()
```

```python
# import nyc vehicle idling data
nyc_vehicle_idling = gpd.read_file('nyc_area_vehicleidling.geojson')
nyc_vehicle_idling.columns
```

```python
# normalize and plot the choropleth map with nyc vehicle idling data
fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(1, 1, 1)
img = nyc_vehicle_idling.plot(ax=ax, column='count', cmap=plt.get_cmap('viridis'))
ax.set_title('NYC vehicle idling')

from matplotlib.colors import Normalize
from matplotlib import cm
norm = Normalize(vmin=nyc_vehicle_idling['count'].min(), vmax=nyc_vehicle_idling['count'].max())
n_cmap = cm.ScalarMappable(norm=norm)
n_cmap.set_array([])
fig.colorbar(n_cmap)
```
